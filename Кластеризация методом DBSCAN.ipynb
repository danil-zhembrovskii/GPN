```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5, metric = 'euclidean')
dbscan.fit(df)
dbscan.labels_    #массив номеров кластеров 
```

```python
a = np.unique(dbscan.labels_, return_counts = True)  # определяем уникальные значения, получается 2 массива: 
# 1 - номера кластеров, 2 - количество элементов в них
a
```

```python
np.asarray(a).T # превращаю переменную а, содержащую 2 массива в единый массив и транспонирую его
# получается избыточное количество кластеров
```

```python
# увеличиваю радиус eps
dbscan_1 = DBSCAN(eps = 3, min_samples = 4, metric = 'euclidean')
dbscan_1.fit(df)
dbscan_1.labels_
np.asarray(np.unique(dbscan_1.labels_, return_counts = True)).T

# слишком высокие значения, поэтому получается 1 кластер
```

```python
eps = np.arange(1, 3, 0.1) # массив от 1 до 3 с промежутком 0.1 (т.к. range работает только с int)
points = np.arange(2, 5, 1)
max_sil = 0 
for i in eps:
    for j in points:
        dbscan = DBSCAN(eps = float(i), min_samples = float(j), metric = 'euclidean') # так как числа небольшие, 
        # то можно не преобразовывать их из numpy.float64 в int
        dbscan.fit(df)
        if len(np.unique(dbscan.labels_)) != 1: # поставил ограничение, так как могут образовываться массивы с 1 кластером
            # а это вызывает ошибку при просчёте силуэта
            sil = metrics.silhouette_score(df, dbscan.labels_)
            if max_sil <= sil:   # поставил такое условие, чтобы значение переменной max_sil достигло максимума, а затем выбрать
                #параметры dbscan с максимальным значением силуэт
                max_sil = sil
                print(np.asarray(np.unique(dbscan.labels_, return_counts = True)))
```

#### Результаты неудовлетворительные. Получается 20 кластеров. Силуэт не репрезентативен, значение данного критерия высокое, однакополучается слишком много кластеров

```python
for i in eps:
    for j in points:
        comb = []
        dbscan = DBSCAN(eps = i, min_samples = j, metric = 'euclidean')
        dbscan.fit(df)
        if 4 <= (len(np.unique(dbscan.labels_))) <= 6:  
            comb.append(round(i,1))
            comb.append(round(j, 1))
            print(np.asarray(np.unique(dbscan.labels_, return_counts = True)))
            print(comb)
# границы количества кластеров подобраны эмпирически: нижняя - количество кластеров при ICA (читерство во благо), 
# верхняя - предположение о возможном наличии 2 дополнительных группах уникумов, так или иначе отличающихся от остальных
```

Методом DBSCAN 4 кластера не получаются с адекватным количеством респондентов.
Буду рассматривать удачные кластеризации, где получилось 6 кластеров. Значения eps и min_samples возьму из списка comb

```python
dbscan_2 = DBSCAN(eps = 1.8, min_samples = 2, metric = 'euclidean')
dbscan_2.fit(df)
dbscan_2.labels_
np.asarray(np.unique(dbscan_2.labels_, return_counts = True))

# читерство, но честное. Сравнивая с результатами, полученными ICA и Kmeans, это самый близкий по количеству респондентов 
# в кластерах, хотя добавляется 2 лишних кластера
```

```python
# реализую иерархический кластерный анализ без построения дендрограммы
link = linkage(df, 'ward', 'euclidean')
link[:10]
```

```python
# строим таблицу сопряженности
cross = pd.DataFrame()
cross['ICA'] = fcluster(link, 4, criterion = 'maxclust')
cross['DBSCAN'] = dbscan_2.labels_

tab = pd.crosstab(cross['ICA'], cross['DBSCAN'])
tab
```

```python
df_2 = df.copy()
df_2['cluster'] = fcluster(link, 4, criterion = 'maxclust')
df_2.groupby(df_2['cluster']).mean()
```

```python
df_3 = df.copy()
df_3['cluster'] = dbscan_2.labels_
df_3.groupby(df_3['cluster']).mean()
```
